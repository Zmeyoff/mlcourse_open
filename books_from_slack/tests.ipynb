```python
def remove_duplicates(lst):
    new_lst = list(set(lst))
    return new_lst
```

```python
remove_duplicates([1,2,1,3])
```

```python
def median(lst):
    new_lst = sorted(lst)
    if len(lst)%2 == 0:
        return((new_lst[len(lst)//2-1]+new_lst[len(lst)//2])/2.0)
    else:
        return(new_lst[len(lst)//2])
```

```python
median([1,2,3,4,5])
```
